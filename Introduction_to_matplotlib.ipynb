```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.tail()
```

```python
heart_disease["age"].plot.hist(bins=100);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50.head()
```

Plot Method

```python


over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target', figsize=(10, 6));
```

```python
# 00 Method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x='age',
             y='chol',
             c='target',
             ax=ax);
#ax.set_ylim([100, 600])
```

```python
over_50.target.values
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex= True)


scatter1 = ax1.scatter(x=over_50["age"],
                     y=over_50["chol"],
                    c=over_50['target'],
                      cmap='winter')

ax1.set_xlim([50, 80])
# customize the the plot
ax1.set(title="Heart Disease and Cholesterol levels",
      ylabel='Cholestrol');

#Add a legend
ax1.legend(*scatter1.legend_elements(), title="Target");

# Add a horizontal line

ax1.axhline(over_50['chol'].mean(), c='g', linestyle='-.');


scatter2 = ax2.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                       cmap='winter')
ax2.set_xlim([50, 80])
ax2.set_ylim([60, 200])
ax2.set(title=" Heart Disease and Max Heart Rate Levels",
        xlabel='Age',
        ylabel='Thalach');
ax2.legend(*scatter2.legend_elements(), title="Target");

ax2.axhline(over_50['thalach'].mean(), linestyle='-.', c='r');

fig.suptitle('Heart Disease Analysis', fontsize= 16, fontweight='bold');
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex= True)


scatter1 = ax1.scatter(x=over_50["age"],
                     y=over_50["chol"],
                    c=over_50['target'],
                      cmap='winter')

ax1.set_xlim([50, 80])
# customize the the plot
ax1.set(title="Heart Disease and Cholesterol levels",
      ylabel='Cholestrol');

#Add a legend
ax1.legend(*scatter1.legend_elements(), title="Target");

# Add a horizontal line

ax1.axhline(over_50['chol'].mean(), c='g', linestyle='-.');


scatter2 = ax2.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                       cmap='winter')
ax2.set_xlim([50, 80])
ax2.set_ylim([60, 200])
ax2.set(title=" Heart Disease and Max Heart Rate Levels",
        xlabel='Age',
        ylabel='Thalach');
ax2.legend(*scatter2.legend_elements(), title="Target");

ax2.axhline(over_50['thalach'].mean(), linestyle='-.', c='r');

fig.suptitle('Heart Disease Analysis', fontsize= 16, fontweight='bold');
```
